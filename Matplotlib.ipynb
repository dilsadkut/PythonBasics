```
x = np.array([10, 30, 40, 20])
y = np.array(["R", "Julia", "Python", "Java"])
colors = ["blue", "orange", "red", "grey"]
plt.pie(x, labels=y, colors=colors)
plt.show()
```

```
plt.pie(x, labels=y, colors=colors, radius=1.5)
plt.show()
```

```
plt.pie(x, labels=y, 
        colors=colors,
        radius=1.5,
        explode=(0, 0, 0.2, 0)
        )
plt.show()
```

```
plt.pie(x, labels=y, 
        colors=colors,
        radius=1.5,
        explode=(0, 0, 0.2, 0),
        shadow=True
        )
plt.show()
```

```
plt.pie(x, labels=y, 
        colors=colors,
        radius=1.5,
        explode=(0, 0, 0.2, 0),
        shadow=True,
        autopct="%1.2f%%"
        )
plt.show()
```

```
x1 = np.arange(0, 20, 0.1)
x2 = np.arange(0, 20, 0.2)

plt.figure()

plt.subplot(211)
plt.plot(x1, np.cos(x1))

plt.subplot(212)
plt.plot(x2, np.sin(x2), 'r-.')
plt.grid()
plt.show()
```

```
x1 = np.arange(0, 20, 0.1)
x2 = np.arange(0, 20, 0.2)

plt.figure(figsize=(10, 5), dpi=300)

# plt.subplot(nrows, ncols, index)
plt.subplot(221)
plt.plot(x1, np.cos(x1))

plt.subplot(222)
plt.plot(x2, np.sin(x2), 'r-.')
plt.grid()
plt.show()
```

```
# https://moonbooks.org/Articles/How-to-plot-a-normal-distribution-with-matplotlib-in-python-/
import scipy.stats

x_min = 0.0
x_max = 16.0

mean = 8.0 
std = 2.0

x = np.linspace(x_min, x_max, 100)

# probability density function (PDF)
y = scipy.stats.norm.pdf(x, mean, std)

plt.plot(x, y, color='coral')

plt.grid()

plt.xlim(x_min, x_max)
plt.ylim(0, 0.25)

plt.title('How to plot a normal distribution in python with matplotlib',fontsize=10)

plt.xlabel('x')
plt.ylabel('Normal Distribution')

plt.savefig("normal_distribution.png")
plt.show()
```

```
x = np.array([1, 2, 3, 4, 5, 6, 7])
y = x ** 2

plt.figure(figsize=(10, 5), dpi=300)

plt.subplot(311)
plt.scatter(x, y)
plt.title("Scatter plot")

plt.subplot(312)
plt.plot(x, y, 'r--', linewidth=5)
plt.grid()
plt.title("Line plot")

plt.subplot(313)
plt.bar(x, y)
plt.title("Bar plot")

plt.suptitle("All the plots")
# plt.tight_layout(pad=1.0)
plt.show()
```
